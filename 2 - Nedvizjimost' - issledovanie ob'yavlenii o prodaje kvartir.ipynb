## Предобработка данных

```python
#выявляем долю пропущенных и неопределённых значений
df.isna().mean().sort_values(ascending = False)
```

```python
#заполняем пропуски
df['is_apartment'] = df['is_apartment'].fillna(False)
df.dropna(subset= ['floors_total', 'locality_name'], inplace = True)
df['balcony'] = df['balcony'].fillna(0)
df['citycenters_nearest'] = df['citycenters_nearest'].fillna('неизвестно')
df.isna().mean().sort_values(ascending = False)
```

```python
#меняем тип данных
df[['last_price','floors_total', 'balcony','last_price', 'rooms']] = df[['last_price','floors_total', 'balcony', 'last_price', 'rooms']].astype('int')
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
cols = ['ceiling_height', 'kitchen_area', 'living_area', 'kitchen_area', 'airports_nearest', 'citycenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
#Создаём функцию для группировки по общей площади квартир
def group_area(total_area):
  if total_area >= 12 and total_area <= 30: #от минимального значения до пика
    return '12 - 30m'
  elif total_area > 30 and total_area <= 50:
    return '30 - 50m'
  elif total_area > 50 and total_area <= 70:
    return '50 - 70m'
  elif total_area > 70 and total_area <= 110:
    return '70 - 110m'
  elif total_area > 110 and total_area <= 200:
    return '110 - 200m'
  return '> 200m'

df['group_area'] = df['total_area'].apply(group_area)
df.head()
```

```python
#определяем количество уникальных значений в столбце с группировкой по площадям
types_area = df['group_area'].unique()

#создаём цикл для заполнения площади кухни медианными значениями по группам
for i in types_area:
  area = df.loc[df['group_area'] == i]
  median_kitchen = area['kitchen_area'].median()
  df.loc[(df['group_area'] == i) & (df['kitchen_area'].isnull()), 'kitchen_area'] = median_kitchen

#создаём таблицу с данными только по площадям
ratio = df[['total_area', 'living_area', 'kitchen_area', 'group_area', 'ceiling_height']]
ratio.head()
```

```python
#создаём цикл для заполнения общей площади медианными значениями по группам
for i in types_area:
  area = df.loc[df['group_area'] == i]
  median_living = area['living_area'].median()
  df.loc[(df['group_area'] == i) & (df['living_area'].isnull()), 'living_area'] = median_living
ratio.head(10)
```

```python
#создаём цикл для заполнения высоты потолков медианными значениями по группам
for i in types_area:
  area = df.loc[df['group_area'] == i]
  median_ceiling = area['ceiling_height'].median()
  df.loc[(df['group_area'] == i) & (df['ceiling_height'].isnull()), 'ceiling_height'] = median_ceiling
ratio.head(10)
```

## Рассчёты и добавление результатов в таблицу

#### Расчёт стоимости квадратного метра

```python
df['price_m'] = (df['last_price'] / df['total_area']).astype('int')
df.head()
```

#### Определение категории этажа

```python
def which_floor(floor):
  if floor == 'floors_total':
    return 'последний'
  elif floor == 1:
    return 'первый'
  return 'другой'
df['which_floor'] = df['floor'].apply(which_floor)
df.head()
```

#### Определяем день недели, месяц и год публикации объявления

```python
df['weekday'] = df['first_day_exposition'].dt.day_name()
df['month'] = df['first_day_exposition'].dt.month_name()
df['year'] = df['first_day_exposition'].dt.year

df.head()
```

#### Расчёт соотношения площадей

```python
#рассчитаем соотношение жилой площади к общей 
df['area1'] = df['living_area'] / df['total_area']
#соотношение площади кухни к общей
df['area2'] = df['kitchen_area'] / df['total_area']
df.head()
```
